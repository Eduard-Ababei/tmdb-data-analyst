```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import sys

# Extend Python path to enable project modules
BASE_PATH = Path('..').resolve().parent
sys.path.append(str(BASE_PATH))

from src.analytics import product_metrics as pm
print("Environment initialized.")
```

## 1. Top Genres by Movie Count
This metric helps understand **catalog concentration** and genre distribution.

```python
df_genres = pm.top_genres()
df_genres.head()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df_genres, x='genre_name', y='movie_count')
plt.xticks(rotation=45)
plt.title('Top Genres by Movie Count')
plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.show()
```

## 2. Average Popularity per Genre
Provides insight into **viewer interest** and high‑performing content categories.

```python
df_pop = pm.popularity_by_genre()
df_pop.head()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df_pop, x='genre', y='avg_popularity')
plt.xticks(rotation=45)
plt.title('Average Popularity per Genre')
plt.xlabel('Genre')
plt.ylabel('Avg Popularity')
plt.show()
```

## 3. Popularity Trend by Release Year
Shows how **audience attention evolves over time** across the entire catalog.

```python
df_trend = pm.popularity_trend()
df_trend.head()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=df_trend, x='year', y='avg_popularity')
plt.title('Popularity Trend Over Time')
plt.xlabel('Release Year')
plt.ylabel('Avg Popularity')
plt.show()
```

## 4. Engagement Score (vote_count × popularity)
A combined metric capturing **interaction intensity** and **content visibility**.
Useful for identifying **high‑impact titles**.

```python
df_engagement = pm.engagement_score()
df_engagement.head(20)
```

## 5. Catalog Maturity (Movies per Year)
Indicates how the catalog has grown over time and highlights content availability patterns.

```python
df_maturity = pm.catalog_maturity()
df_maturity.head()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=df_maturity, x='year', y='total_movies')
plt.title('Catalog Maturity Over Time')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.show()
```

## 6. Genre Stability (Variance in Popularity)
Low variance = consistent performance.
High variance = some movies perform strongly, others poorly.
A key metric for understanding **content predictability**.

```python
df_stability = pm.genre_stability()
df_stability.head()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df_stability, x='genre', y='popularity_variance')
plt.xticks(rotation=45)
plt.title('Genre Stability (Popularity Variance)')
plt.xlabel('Genre')
plt.ylabel('Popularity Variance')
plt.show()
```
